## Euler Discretization

$$
S_t = S_{t-1} \cdot exp((r - 0.5 \cdot stdev^2) \cdot delta_t + stdev \cdot delta_t^{0.5} \cdot Z_t)
$$

```python
r = 0.025
T = 1.0 
t_intervals = 1000
delta_t = T / t_intervals  
iterations = 10000  
```

```python
Z = np.random.standard_normal((t_intervals + 1, iterations))  
S = np.zeros_like(Z) 
S0 = data.iloc[-1]  
S[0] = S0 
range = np.arange(1, t_intervals+1)
for t in range:
    S[t] = S[t-1] * np.exp((r - 0.5 * stdev ** 2) * delta_t + stdev * delta_t ** 0.5 * Z[t])
```

```python
S
```

```python
S.shape
```

$$
C = \frac{exp(-r \cdot T) \cdot \sum{p_i}}{iterations}
$$

Forecast the price of a call stock option using the above formula.

```python
p = np.maximum(S[-1] - S0, 0)
p
```

```python
p.shape
```

```python
np.sum(p)
```

```python
C = np.exp(-r * T) * np.sum(p) / iterations
C  
```

## Black-Scholes-Merton theorem

$$
d_1 = \frac{\ln(\frac{S}{K}) + (r + \frac{stdev^2}{2})t}{s \cdot \sqrt{t}}
$$

$$
d_2 = d_1 - s \cdot \sqrt{t} = \frac{\ln(\frac{S}{K}) + (r - \frac{stdev^2}{2})t}{s \cdot \sqrt{t}}
$$

Made user defined functions for d1 & d2. 

```python
def d1(S, K, r, stdev, T):
    return (np.log(S / K) + (r + stdev ** 2 / 2) * T) / (stdev * np.sqrt(T))
 
def d2(S, K, r, stdev, T):
    return (np.log(S / K) + (r - stdev ** 2 / 2) * T) / (stdev * np.sqrt(T))
```

```python
norm.cdf(0), norm.cdf(0.25), norm.cdf(0.75), norm.cdf(9)
```

$$
\textbf{C} = SN(d_1) - Ke^{-rt}N(d_2) 
$$

Made User Defined function for calculating Call price.

```python
def BSM(S, K, r, stdev, T):
        return (S * norm.cdf(d1(S, K, r, stdev, T))) - (K * np.exp(-r * T) * norm.cdf(d2(S, K, r, stdev, T)))
```
